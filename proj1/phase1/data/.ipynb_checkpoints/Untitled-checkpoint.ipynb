```python
isu = pd.read_csv('isu.csv')
npinput = isu.to_numpy()
nparr = np.zeros((int(npinput.shape[0]/5),2))
total = 0
for i in range (0, int(npinput.shape[0]/5)):
    for j in range (0, 5):
        total = total + npinput[(i*5) + j][1] * 100
    nparr[i] =  (total/5, npinput[i*5][0])
plt.loglog(nparr[:, 0], nparr[:, 1], label="Insertion Sort Random")

isa = pd.read_csv('isa.csv')
npinput = isa.to_numpy()
nparr = np.zeros((int(npinput.shape[0]/5),2))
total = 0
for i in range (0, int(npinput.shape[0]/5)):
    for j in range (0, 5):
        total = total + npinput[(i*5) + j][1] * 100
    nparr[i] =  (total/5, npinput[i*5][0])
plt.loglog(nparr[:, 0], nparr[:, 1], label='Insertion Sort Almost Sorted')

isr = pd.read_csv('isr.csv')
npinput = isr.to_numpy()
nparr = np.zeros((int(npinput.shape[0]/5),2))
total = 0
for i in range (0, int(npinput.shape[0]/5)):
    for j in range (0, 5):
        total = total + npinput[(i*5) + j][1] * 100
    nparr[i] =  (total/5, npinput[i*5][0])
plt.loglog(nparr[:, 0], nparr[:, 1], label='Insertion Sort Reverse')

msu = pd.read_csv('msu.csv')
npinput = msu.to_numpy()
nparr = np.zeros((int(npinput.shape[0]/5),2))
total = 0
for i in range (0, int(npinput.shape[0]/5)):
    for j in range (0, 5):
        total = total + npinput[(i*5) + j][1] * 100
    nparr[i] =  (total/5, npinput[i*5][0])
plt.loglog(nparr[:, 0], nparr[:, 1], label="Merge Sort Random")

msa = pd.read_csv('msa.csv')
npinput = msa.to_numpy()
nparr = np.zeros((int(npinput.shape[0]/5),2))
total = 0
for i in range (0, int(npinput.shape[0]/5)):
    for j in range (0, 5):
        total = total + npinput[(i*5) + j][1] * 100
    nparr[i] =  (total/5, npinput[i*5][0])
plt.loglog(nparr[:, 0], nparr[:, 1], label='Merge Sort Almost Sorted')

msr = pd.read_csv('msr.csv')
npinput = msr.to_numpy()
nparr = np.zeros((int(npinput.shape[0]/5),2))
total = 0
for i in range (0, int(npinput.shape[0]/5)):
    for j in range (0, 5):
        total = total + npinput[(i*5) + j][1] * 100
    nparr[i] =  (total/5, npinput[i*5][0])
plt.loglog(nparr[:, 0], nparr[:, 1], label='Merge Sort Reverse')

plt.legend()
#plt.figure(figsize=(30,20))

plt.show()
```
